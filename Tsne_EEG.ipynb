```python
from scipy.io import loadmat

filepath = 'Data/SampleData_EEGLearn/' # Location of Data
file='FeatureMat_timeWin.mat'  #file name

xx = loadmat(filepath+file)
    
data = xx['features']

data_x=data[:,0:-1]
data_y=data[:,-1]

print(data_x.shape, '==> Data(trials,Features)')

print(data_y.shape, '==> Data(trials,)')
```

```python
## Run Tsne
import numpy as np
from sklearn.manifold import TSNE

X_embedded = TSNE(perplexity=30, n_components=3,init='pca', n_iter=3000).fit_transform(data_x)

X_embedded.shape
```

```python
#2D Plot
%matplotlib notebook

import matplotlib.pyplot as plt

plt.clf()

plt.scatter(X_embedded[:, 0],X_embedded[:, 2], 15, data_y)
plt.show()
```

```python
#2D Plot
%matplotlib notebook

import matplotlib.pyplot as plt

plt.clf()

plt.scatter(X_embedded[:, 0],X_embedded[:, 1], 15, data_y)
plt.show()
```

```python
#%matplotlib notebook

import ipyvolume.pylab as ipv


from skimage import color
label=color.label2rgb(data_y)

ipv.scatter(X_embedded[:, 0],X_embedded[:, 1], X_embedded[:, 2],size=3, marker="sphere",color=label)

ipv.ylim(-40,40)
ipv.zlim(-40,40)


ipv.show()
```
